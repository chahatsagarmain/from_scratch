```python
import linear_regression
import matplotlib.pyplot as plt 
import numpy as np
```

```python
X = np.array([i for i in range(0,10)])
y = np.array([i for i in range(0,20,2)])
print(len(X))
print(len(y))
```

```python
fig , ax = plt.subplots()

ax.scatter(X,y,color='red')
```

```python
model = linear_regression.LinearRegression()
model.train(X,y,0.01,10000)
```

```python
print(type(model))
```

```python
plt.scatter(X,y , color = 'red')
plt.plot(X , model.predict(X))
print(model.predict(X))
print(y)
```
